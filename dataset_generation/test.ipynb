```python
import numpy as np
from ipycanvas import Canvas, hold_canvas
from IPython.display import display
from PIL import Image

def blob(x, y, r):
    x1 = x
    x2 = np.random.randint(x - 40, x + 40)
    x3 = np.random.randint(x - 40, x + 40)
    x4 = np.random.randint(x - 40, x + 40)

    y1 = y
    y2 = np.random.randint(y - 40, y + 40)
    y3 = np.random.randint(y - 40, y + 40)
    y4 = np.random.randint(y - 40, y + 40)

    r1 = r
    r2 = np.random.randint(r - 10, r + 10)
    r3 = np.random.randint(r - 10, r + 10)
    r4 = np.random.randint(r - 10, r + 10)
    
    lst = [[x1, y1, r1], [x2, y2, r2], [x3, y3, r3], [x4, y4, r4]]

    return lst

#def save_to_file(idx, canvas):
#    canvas.to_file(fr"C:\Users\magfa\Documents\Master\Masteroppgave\synthetic_dataset\label\{idx}.png")

images = []

def draw_and_save(n_images):
    canvas = Canvas(width=256, height=179, sync_image_data = True)
    #canvas.observe(save_to_file, "image_data")
    for i in range(n_images):
        canvas.clear()
        n_circles = np.random.randint(1, 10)
        n_plates = np.random.randint(2, 16)
        x_circle = np.random.randint(0, canvas.width, size=(n_circles))
        x_circle = x_circle.tolist()
        print(len(x_circle))
        y_circle = np.random.randint(0, canvas.height, size=(n_circles))
        y_circle = y_circle.tolist()
        r = np.random.randint(10, 20, size=(n_circles))
        r = r.tolist()

        x_plates = np.random.randint(0, canvas.width, size=(n_plates))
        y_plates = np.random.randint(0, canvas.height, size=(n_plates))
        if len(x_plates) % 2 != 0:
            x_plates = x_plates[:-1]
            y_plates = y_plates[:-1]
        pos_plates = tuple(zip(x_plates, y_plates))
        pos_plates = [(int(x), int(y)) for x, y in zip(x_plates, y_plates)]

        n_arcs = np.random.randint(10, 30)
        x_arcs = np.random.randint(0, canvas.width, size=(n_arcs))
        y_arcs = np.random.randint(0, canvas.height, size=(n_arcs))
        r_arcs = np.random.randint(10, 150, size=(n_arcs))
        start_angle = np.random.randint(0, 360, size=(n_arcs))
        end_angle = start_angle + 0.05

        new_x_circle, new_y_circle, new_r = [], [], []
        for k in range(len(x_circle)):
            positions = blob(x_circle[k], y_circle[k], r[k])
            for l in range(len(positions)):
                new_x_circle.append(positions[l][0])
                new_y_circle.append(positions[l][1])
                new_r.append(positions[l][2])
        print(len(new_x_circle))


        canvas.fill_style = 'blue'
        canvas.fill_arcs(x_arcs, y_arcs, r_arcs, start_angle, end_angle)
        canvas.stroke_style = "blue"
        canvas.line_width = 2
        for j in range(int(len(x_plates) / 2)):
            canvas.stroke_line(int(pos_plates[j][0]), int(pos_plates[j][1]), int(pos_plates[j + int(len(x_plates) / 2)][0]), int(pos_plates[j + int(len(x_plates) / 2)][1]))
        canvas.fill_style = 'red'
        canvas.fill_circles(new_x_circle, new_y_circle, new_r)

        display(canvas)
        images.append(canvas.get_image_data())


draw_and_save(3)
print(len(images))
```
